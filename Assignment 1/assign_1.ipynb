Data Cleaning

Drop the row whose column = Customer_ID is Missing

```python
df.dropna(subset=['Customer ID'], inplace=True)
```

```python
df.isnull().sum()
```

Add mode value to the row whose Product Category is missing

```python
mode_value = df['Product Category'].mode()[0]
```

```python
df['Product Category'].fillna(mode_value, inplace=True)
```

```python
df.isnull().sum()
```

Add values using median on the row whose Column 'Quantity' and 'Price per Unit' is missing

```python
df['Quantity'].fillna(df['Quantity'].median(), inplace=True)
df['Price per Unit'].fillna(df['Price per Unit'].median(), inplace=True)
```

```python
df.isnull().sum()
```

Add values using multiplication of column 'Quantity' and 'Price per Unit' if 'Total Amount' column is empty.

```python
df['Total Amount'].fillna(df['Quantity'] * df['Price per Unit'], inplace=True)
```

```python
df.isnull().sum()
```

Remove time from date column

```python
df['Date'] = pd.to_datetime(df['Date']).dt.date
```
